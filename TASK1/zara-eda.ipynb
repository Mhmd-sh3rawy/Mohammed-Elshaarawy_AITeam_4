```python
import numpy as np 
import matplotlib.pyplot as plt
import opendatasets as od 
import pandas as pd 
```

```python
dataset_url = 'https://www.kaggle.com/datasets/marixe/zara-sales-for-eda'
od.download(dataset_url)
```

```python
df = pd.read_csv("zara-sales-for-eda/Zara_sales_EDA.csv",sep=";")
display(df.head())
display(df.tail())
```

```python
display(df.isna().sum())
display(df.info())
display(df.describe())
```

```python
display(df[df.isna()["description"]])
display(df[df.isna()["name"]])
```

```python
cat_labels = ["Product Position", "Promotion", "Product Category", "Seasonal", "brand", "currency", "terms", "section", "season", "material", "origin"]

for label in cat_labels:
    display(df[label].value_counts())
```

```python
print("WOMAN Sales Volume : ", df[df["section"] == "WOMAN"]["Sales Volume"].sum())
print("MAN Sales Volume : ", df[df["section"] == "MAN"]["Sales Volume"].sum())
print(f"WOMAN Sales Volume / MAN Sales Volume =  \n ", df[df["section"] == "WOMAN"]["Sales Volume"].sum() / df[df["section"] == "MAN"]["Sales Volume"].sum())
```

```python
woman_section_describe = df[df["section"] == "WOMAN"][["Sales Volume", "price"]].describe()
woman_section_describe
```

```python
man_section_describe = df[df["section"] == "MAN"][["Sales Volume","price"]].describe() 
man_section_describe
```

```python
df_copy = df.copy()
df_copy["Total sales"] = df_copy["price"] * df["Sales Volume"] 
print(df_copy[df_copy["section"] == "WOMAN"]["Total sales"].sum(), " Mean " , df_copy[df_copy["section"] == "WOMAN"]["Total sales"].mean())
print(df_copy[df_copy["section"] == "MAN"]["Total sales"].sum(), " Mean ", df_copy[df_copy["section"] == "MAN"]["Total sales"].mean())
print(df_copy[df_copy["section"] == "WOMAN"]["Total sales"].sum() / df_copy[df_copy["section"] == "MAN"]["Total sales"].sum())
```
